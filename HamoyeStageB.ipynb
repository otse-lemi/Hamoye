```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.metrics import mean_squared_error
```

```
energy_data = pd.read_csv("energydata_complete.csv")
energy_data
```

```
energy_data.isna().sum()
```

```
X = pd.DataFrame(energy_data['T2'])
y = pd.DataFrame(energy_data['T6'])
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

linear_model = LinearRegression()

#fit the model to the training dataset
linear_model.fit(x_train, y_train)
linear_model.score(x_test,y_test)
predict_linear_model = linear_model.predict(x_test)
predicted_values = linear_model.predict(x_test)


print('r2 score on training set:', linear_model.score(x_train,y_train))
print('r2 score on test set:', linear_model.score(x_test,y_test))
#print(predicted_values)

r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
print(r2_score)
print('Mean Absolute Error on predicted value:', mean_absolute_error(y_test,predict_linear_model))
```

```
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
print(rss)
```

```
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
print(rmse)
```

```
scaler = MinMaxScaler()
features = energy_data.drop(columns=['date', 'lights'])
standardised_df = pd.DataFrame(scaler.fit_transform(features), columns=features.columns)
target = standardised_df['T6']

#Split the dataset
x_train, x_test, y_train, y_test = train_test_split(standardised_df.drop('T6', axis=1), target, test_size=0.3, random_state=42)

#model
linear_model = LinearRegression()

#fit the model to the training dataset
linear_model.fit(x_train, y_train)
linear_model.score(x_test,y_test)
predict_linear_model = linear_model.predict(x_test)

print('r2 score on training set:', linear_model.score(x_train,y_train))
print('r2 score on test set:', linear_model.score(x_test,y_test))
print('Mean Absolute Error on predicted value:', mean_absolute_error(y_test,predict_linear_model))
```
